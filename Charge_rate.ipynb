```python
import matplotlib.pyplot as plt
import numpy as np

# Battery parameters
battery_capacity = 50  # kWh
start_percent = 10
end_percent = 80
charging_rate = 7

# Convert percentages to kWh
start_kwh = battery_capacity * start_percent / 100
end_kwh = battery_capacity * end_percent / 100
charged_kwh = end_kwh - start_kwh

# Time to charge in hours
charging_time_hours = charged_kwh / charging_rate

# Generate time and energy data
time = np.linspace(0, charging_time_hours, 100)
energy = start_kwh + charging_rate * time

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(time, energy, label='Charging Curve', color='blue')
plt.title('Charging Rate vs Time')
plt.xlabel('Time (hours)')
plt.ylabel('Energy (kWh)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
